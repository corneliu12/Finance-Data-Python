## Normalizing Time Series to a Base Value (100)

```python
close.head()
```

```python
# value of the first value (row 0, column 0)
close.iloc[0,0]
```

```python
# dividing AAPL close by the first value
close.AAPL.div(close.iloc[0,0]).mul(100)
```

```python
# the first row of the dataframe
close.iloc[0]
```

```python
# normalizing the values by dividing each row 
# by the values of the first row and muliply by 100
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize = (15,8), fontsize = 14)
plt.legend(fontsize = 14)
plt.title("Normalized Prices", fontsize = 14)
plt.show()
```

```python
std_daily_Returns = np.sqrt(var_daily_Returns)
std_daily_Returns
```

```python
ann_mean_Return = ret.mean() * 252
ann_mean_Return
```

```python
ann_var_Returns = ret.var() * 252
ann_var_Returns
```

```python
ann_std_Returns = np.sqrt(ann_var_Returns)
ann_std_Returns
```

## Financial Times Series - Return and Risk

### Comparing Risk and Return

```python
norm.plot(figsize = (15, 8), fontsize = 14)
plt.legend(fontsize = 13)
plt.show()
```

```python
close.head()
```

```python
ret = close.pct_change().dropna()
```

```python
ret.head()
```

```python
# transpose and select mean and std for the describe
summary = ret.describe().T.loc[:, ["mean", "std"]]
summary
```

```python
summary["mean"] = summary["mean"] * 252
summary["std"] = summary["std"] * np.sqrt(252)
```

```python
# the average yearly percent of change since 2010 to Sept 2020
# and the average yearly standard deviation for the same period
summary
```

```python
summary.plot.scatter(x = "std", y = "mean", figsize = (12, 8), s=50, fontsize = 15)
for i in summary.index:
    plt.annotate(i, xy=(summary.loc[i, "std"] + 0.002, summary.loc[i, "mean"] + 0.002), size = 15)
plt.xlabel("ann. Risk(std)", fontsize = 15)
plt.ylabel("ann. Return", fontsize = 15)
plt.title('Risk/Return', fontsize = 20)
plt.show()
```
