## Normalizing Time Series to a Base Value (100)

```python
close.head()
```

```python
# value of the first value (row 0, column 0)
close.iloc[0,0]
```

```python
# dividing AAPL close by the first value
close.AAPL.div(close.iloc[0,0]).mul(100)
```

```python
# the first row of the dataframe
close.iloc[0]
```

```python
# normalizing the values by dividing each row 
# by the values of the first row and muliply by 100
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.plot(figsize = (15,8), fontsize = 14)
plt.legend(fontsize = 14)
plt.title("Normalized Prices", fontsize = 14)
plt.show()
```

## The shift() method

```python
close.head()
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
# shifts the index by a number of periods
aapl.shift(periods = 1)
```

```python
aapl['lag1'] = aapl.shift(periods = 1)
```

```python
aapl.head()
```

```python
aapl.AAPL.sub(aapl.lag1)
```

```python
aapl['Diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl['pct_change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

## The methods diff() and pct_change()

```python
aapl.head()
```

```python
# calculating change from day to day using diff() method
aapl.AAPL.diff(periods = 1)
```

```python
aapl['Diff2'] = aapl.AAPL.diff(periods = 1)
```

```python
aapl.head(10)
```

```python
aapl["pct_change2"] = aapl.AAPL.pct_change(periods = 1).mul(100)
```

```python
aapl.head(10)
```

```python
# resample to monthly prices using last business day of the month
aapl.AAPL.resample("BM").last().pct_change(periods = 1).mul(100)
```

```python
# resample to weekly prices 
aapl.AAPL.resample("W").last().pct_change(periods = 1).mul(100)
```
